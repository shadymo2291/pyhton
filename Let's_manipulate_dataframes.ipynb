```python
# 6 - Add a new column "Success": if the score is higher than 10 we will have 1, else we will have 0.
exam_data = {'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],

'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],    # nan stands for not a number

'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],

'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']}

labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']

import numpy as np
import pandas as pd
myva = pd.DataFrame(exam_data, index = labels)

score_list = myva["score"].tolist()
Success_list = []
for i in score_list:
  if i > 10:
    Success_list.append(1)
  else:
    Success_list.append(0)
Success = pd.Series(Success_list)

myva["Success"] = Success.values

myva
```
