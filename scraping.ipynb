```python
from requests_html import HTMLSession
from bs4 import BeautifulSoup

ht = HTMLSession() #sending get request
url = "https://en.wikipedia.org/wiki/Egypt"  #my link

def data(x):  #function to parse the link of the page
  gt = ht.get(x)
  soup = BeautifulSoup(gt.text, 'html.parser')
  return soup

soup = data(url)

print(soup)
```

```python
#1.2
def title(soup):
  titles = soup.select_one(".mw-page-title-main")
  return titles.get_text()

title(soup)
```

```python
#1.3
def paragraphs():
   for head in soup.find_all("p"):
     print(head.get_text())
paragraphs()
```

```python
#1.4
def links():
    links = {}
    for link in soup.find_all("a"):
      url = link.get("href", "")
      if "/wiki/" in url:
        links[link.text.strip()] = url
    print(links,"\n")

links()
```

```python
from requests_html import HTMLSession
from bs4 import BeautifulSoup

def scraper(wiki):

  def data(url):
    gt = ht.get(url)
    soup = BeautifulSoup(gt.text, 'html.parser')
    return soup

  soup = data(wiki)

  def title():
    titles = soup.find("title")
    print(titles,"\n")

  def paragraphs():
   for head in soup.find_all("p"):
     print(head.get_text(),"\n")
  
  def links():
    links = {}
    for link in soup.find_all("a"):
      url = link.get("href", "")
      if "/wiki/" in url:
        links[link.text.strip()] = url
    print(links,"\n")
  

  

  print(title(),paragraphs(),links())

scraper("https://en.wikipedia.org/wiki/Egypt")
```
